```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

## Data for 2020

```python
# loading in data for 2020
df_2020 = pd.read_csv('heart_2020_cleaned.csv')

# creating subset for the sake of run time
df_2020_subset = df_2020.head(10000)

# showing dataframe
df_2020.head()
```

```python
# partition the data
# can be replaced with full dataframe
X   = df_2020_subset[['BMI','Smoking','AlcoholDrinking','Stroke','PhysicalHealth','MentalHealth','DiffWalking','Sex',
               'AgeCategory','Diabetic','PhysicalActivity','SleepTime','Asthma','KidneyDisease',
               'SkinCancer']] #get the input features
y   = df_2020_subset['HeartDisease']              #get the target

X_train, X_test, y_train, y_test = train_test_split(X,              #the input features
                                                    y,              #the label
                                                    test_size=0.3,  #set aside 30% of the data as the test set
                                                    random_state=7, #reproduce the results
                                                    stratify=y      #preserve the distribution of the labels
                                                   )
```

```python
X_train = X_train.copy()
X_test  = X_test.copy()
```

```python
#normalize the numeric features
scaler = StandardScaler()
scaler.fit(X_train[['BMI']])

X_train[['BMI']] = scaler.transform(X_train[['BMI']]) #scale the training data
```

```python
X_train.head()
```

```python
#initialize the classifier

svm = SVC() #the default kernel is rbf
svm.fit(X_train, y_train) #fit the data
```

```python
#predict the labels for the test set
y_pred   = svm.predict(X_test)
print('The predicted heart diesease in 2020 is: {}'.format(y_pred))
```

```python
#evaluate the prediction accuracy
print('The accuracy of the model is: {}'.format(svm.score(X_test, y_test)))
```

```python
#setup the parameter grid to try values of c ranging from 1 to 10, and values for gamma ranging from 1 to 5.
params = {'C': np.arange(1, 10),
         'gamma': np.arange(1, 5)}

svm = SVC()

#initialize Grid Search with Cross Validation
svm_grid = GridSearchCV(estimator = svm, param_grid = params, cv=5, return_train_score = True, n_jobs=-1)
svm_grid.fit(X_train, y_train)
```

```python
print(svm_grid.best_params_) #print the best parameters
print(svm_grid.best_score_)  #print the best score
```

```python
#display the cv results for the train and test set
cv_results = pd.DataFrame(svm_grid.cv_results_)
cv_results = cv_results.sort_values('mean_test_score', ascending=False)
cv_results[['mean_train_score', 'std_train_score', 'mean_test_score', 'std_test_score']].head()
```
